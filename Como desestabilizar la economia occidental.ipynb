### Problema actual, La Inflación.

Primero, definamos que es la inflación.
    La inflación, es el aumento promediado, generalizado y sostenido de los precios. Entran muchas categorías en la inflación, desde precios de la canasta básica, energía, hidrocarburos, servicios, vivienda, etc.

La inflación en pocas palabras, causa que el dinero pierda poder sobre el tiempo, y es la razón por la que nuestros abuelos podían comprar 3 terrenos por $1200 pesos, y todavía sobraba para hacer una carne asada como celebración. Y nosotros, si queremos comprar una casa, nos cueste 3 millones, y unos 15 años de nuestra vida acumular el dinero. 

No necesariamente la inflación es mala per se, un índice inflacionario pequeño (menor a 2% anualizado) es bastante bueno, porque permite el crecimiento económico de un país este estable, principalmente porque alimenta el consumismo (Esto es debido, a que, si los precios constantemente están aumentando, es preferible comprar algo hoy, que mañana, y así sucesivamente. Entonces el consumidor aumenta su aportación económica al mercado, movilizando la economía.) y es un indicador que hay crecimiento económico.

Pero cuando está a niveles altos, es muy peligroso, porque daña nuestro poder adquisitivo, y como respuesta, el consumidor promedio cambia sus hábitos de consumo, haciéndose más conservador y ahorrador, lo que afecta al mercado, que causa menor actividad económica, y hasta deflación (disminución de los precios) como respuesta de vender con menores utilidades, o hasta pérdidas, para poder tener dinero.

### El invisible precio de imprimir dinero.

Todo tiene un precio, y básicamente regalar dinero a tus ciudadanos, es lejos de ser la excepción.

Hubo 3 paquetes de estímulo a lo largo de dos años, y cada uno tuvo un costo considerable, hablamos de 2.2, 2.9, 1.9 Trillones respectivamente, lo que suman una exorbitante cantidad de 7 trillones de dólares inyectados a la economía estadounidense.

Para contextualizar cuanto es esta cantidad, observaremos el índice M2.

El M2 engloba M1, que éste engloba: Dinero físico, depósitos bancarios, cheques de viajeros. Y el M2 añade: Cuentas de ahorro, y diversos tipos de cuentas y fondos.

En síntesis, el M2 es todo el dinero que hay en circulación en una economía, observémoslo en los últimos 40 años.

```python
#Cargar dataset WM2NS, modificar ruta de ser necesario

WM2NS= pd.read_csv('C:/Users/Usuario/Downloads/Portfolio/WM2NS.csv')

WM2NS=WM2NS.rename(columns={'DATE' : 'Date', 'WM2NS': 'Value'})

WM2NS['Date']= pd.to_datetime(WM2NS['Date'])

WM2NS= WM2NS.set_index('Date')
```

```python
figWM2NS= px.line(WM2NS, y='Value', title='Money in the US Economy',
    color_discrete_sequence=['darkolivegreen'], template= 'plotly_white',
    labels={'Value': 'Billions of Dollars (Trillions)', 'Date': 'Year'})

figWM2NS.show()
```

Entonces, se puede observar en la gráfica de arriba como hubo un crecimiento considerable desde marzo 2020, pero para poner en perspectiva que tan significativo es inyectarle 7 trillones a una economía, observemos febrero del 2020, tenía una en circulación 15.3 trillones de dólares, esos 7 trillones representan el 46% de esa cantidad, lo que daría que de los 21 trillones que hay en circulación actualmente, el 32.86% de ese dinero se generó en los últimos 2 años. 

En un país que lleva casi 250 años de historia independiente, 33% de su dinero actual en circulación, se generó en tan solo 2 años, let that sink in.

Veamos cuanto tiempo hubiera tomado en cada década desde los 80s, crecer esos 7 trillones.

```python
data= [
    ['80s', 29.3],
    ['90s', 22.75],
    ['00s', 14.9],
    ['10s', 9.9],
    ['20s', 2.25]
]
yrs= pd.DataFrame(data, columns=['Decade', 'Years'])
```

```python
yrsfig= px.histogram(yrs, x= 'Decade', y='Years', title='Years to Grow 7 Trillion the US Economy', 
    color_discrete_sequence=['#283747'], template= 'plotly_white',)

yrsfig.show()

#ver mejores maneras de representar los datos
```

Y como la inflación se rige generalmente por la ley de la Oferta y la Demanda, al haber mucha demanda, en este caso por un superfluo de dinero disponible en todos los ciudadanos, la oferta baja, por ende, el precio debe de subir para estabilizar la balanza.

También recordemos, que había un estímulo si alguien estaba desempleado, este estímulo era de $800 dólares semanales, el cual era una cantidad bastante agradable, de hecho, era superior a los ingresos de muchísimas personas en trabajos base, entonces era más lógico estar desempleado, que estar trabajando y ganar un menor sueldo.

Al suceder eso, había un desabasto de mano de obra en las empresas, y si no hay mano de obra, tienes que subir las compensaciones de tus empleados para mantenerlos, pero subiendo los sueldos, las empresas no van a quedarse con la pérdida, y generalmente, suben los precios de sus productos o servicios.
